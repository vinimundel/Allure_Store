```python

import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = (pd.read_csv(url))
loja2 = (pd.read_csv(url2))
loja3 = (pd.read_csv(url3))
loja4 = (pd.read_csv(url4))

loja.head()

```

```python
import matplotlib.pyplot as plt
import numpy as np

#Calcular o faturamento de cada uma das lojas
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

faturamento_milhoes1 = faturamento_loja1 / 1000000
faturamento_milhoes2 = faturamento_loja2 / 1000000
faturamento_milhoes3 = faturamento_loja3 / 1000000
faturamento_milhoes4 = faturamento_loja4 / 1000000

#Arredondar duas casas
faturamento_milhoes1 = round(faturamento_milhoes1, 2)
faturamento_milhoes2 = round(faturamento_milhoes2, 2)
faturamento_milhoes3 = round(faturamento_milhoes3, 2)
faturamento_milhoes4 = round(faturamento_milhoes4, 2)

#Gerar tabela gráfica dos faturamentos

tabela_faturamento = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento (Milhões R$)': [faturamento_milhoes1, faturamento_milhoes2, faturamento_milhoes3, faturamento_milhoes4]
})

#Exibir a tabela
print('\nTabela de Faturamento:')
display(tabela_faturamento)

#Gráfico de barras para os faturamentos

plt.figure(figsize=(8, 6)) # Ajusta o tamanho da figura
bars = plt.bar(tabela_faturamento['Loja'], tabela_faturamento['Faturamento (Milhões R$)'], color=['blue', 'green', 'red', 'purple'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlabel('Loja')
plt.ylabel('Faturamento (Milhões R$)')
plt.title('Faturamento das Lojas')
plt.ylim(1.0, 1.6)
plt.yticks(np.arange(1.0, 1.61, 0.1))

# Adicionar os valores acima de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.2f}', va='bottom', ha='center')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Combinar os dados das lojas
loja_total = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

# Vendas por categoria
vendas_por_categoria = loja_total.groupby('Categoria do Produto')['Preço'].sum()

# Criar tabela de vendas por categoria e quantidade
tabela_vendas_por_categoria = vendas_por_categoria.reset_index()

#adicionar a quantidade vendida de cada produto na tabela
quantidade_por_categoria = loja_total.groupby('Categoria do Produto')['Preço'].count()
tabela_vendas_por_categoria['Quantidade'] = quantidade_por_categoria.values
tabela_vendas_por_categoria.columns = ['Categoria do Produto', 'Faturamento', 'Quantidade']

# Porcentagem das vendas por categoria
vendas_totais = vendas_por_categoria.sum()
percentual_vendas_por_categoria = (vendas_por_categoria / vendas_totais) * 100
percentual_vendas_por_categoria = percentual_vendas_por_categoria.sort_values(ascending=False)

# Gráfico de Pizza
plt.figure(figsize=(15, 10))
plt.pie(percentual_vendas_por_categoria, startangle=140)
plt.legend(
    labels=(
        percentual_vendas_por_categoria.index
        + ' - '
        + percentual_vendas_por_categoria.round(2).astype(str)
        + '%'
    ),
    loc="best"
)
plt.title('Distribuição Percentual das Vendas por Categoria em Todas as Lojas')
plt.axis('equal')
plt.show()

# Tabela de vendas por local e categoria
tabela_vendas_por_local_categoria = loja_total.groupby(['Local da compra', 'Categoria do Produto'])['Preço'].sum().reset_index()
tabela_vendas_por_local_categoria.columns = ['Local da compra', 'Categoria do Produto', 'Faturamento']

# Faturamento do produto
tabela_faturamento_produto = loja_total.groupby('Produto').agg(
    Faturamento=('Preço', 'sum'),
    Quantidade=('Preço', 'count')
).reset_index()


# --- Resultados ---
print('\nTabela de Vendas por Categoria:')
display(tabela_vendas_por_categoria)

print('\nTabela de Faturamento por Produto:')
display(tabela_faturamento_produto)

print('\nTabela de Vendas por Local da compra e Categoria:')
display(tabela_vendas_por_local_categoria)
```

```python
import pandas as pd

# Calcular a avaliação media
avaliacao_media_loja1 = loja['Avaliação da compra'].mean()
avaliacao_media_loja2 = loja2['Avaliação da compra'].mean()
avaliacao_media_loja3 = loja3['Avaliação da compra'].mean()
avaliacao_media_loja4 = loja4['Avaliação da compra'].mean()

# Display the average ratings
print(f"Média de avaliação Loja 1: {avaliacao_media_loja1:.2f}")
print(f"Média de avaliação Loja 2: {avaliacao_media_loja2:.2f}")
print(f"Média de avaliação Loja 3: {avaliacao_media_loja3:.2f}")
print(f"Média de avaliação Loja 4: {avaliacao_media_loja4:.2f}")
```

```python
import pandas as pd
#padronizar o nome das lojas
lojas = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4
}

# Criar um Loop iterando em cada loja
for nome_loja, dados_loja in lojas.items():
    vendas_por_produto = dados_loja.groupby('Produto').size()
    vendas_por_produto = vendas_por_produto.sort_values(ascending=False)

    # Achar o produto mais e menos vendido de cada loja pela ordem
    produto_mais_vendido = vendas_por_produto.idxmax()
    vendas_maxima = vendas_por_produto.max()
    produto_menos_vendido = vendas_por_produto.idxmin()
    vendas_minima = vendas_por_produto.min()
    # Formatar o print
    print(f"--- {nome_loja} ---")
    print(f"Produto com mais vendas: {produto_mais_vendido} ({vendas_maxima} unidades)")
    print(f"Produto com menos vendas: {produto_menos_vendido} ({vendas_minima} unidades)")
    print("-" * (len(nome_loja) + 6))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Calcular frete medio em cada loja e arredondar
frete_medio_loja1 = loja['Frete'].mean().round(2)
frete_medio_loja2 = loja2['Frete'].mean().round(2)
frete_medio_loja3 = loja3['Frete'].mean().round(2)
frete_medio_loja4 = loja4['Frete'].mean().round(2)

# Criar tabela para visualização
frete_medio_df = pd.DataFrame({
    'Loja': ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Frete Médio': [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]
})

# Mostrar o frete medio
print("Frete Médio por Loja:")
display(frete_medio_df)

# analisar o frete medio por meio de um gráfico de tabela
plt.figure(figsize=(10, 8))
plt.bar(frete_medio_df['Loja'], frete_medio_df['Frete Médio'], color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Loja')
plt.ylabel('Frete Médio (R$)')
plt.ylim(0, 40)
plt.title('Frete Médio por Loja')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# README

## Introdução:
Este notebook contém uma análise exploratória dos dados de vendas de quatro lojas diferentes. O objetivo é entender o desempenho de cada loja, o comportamento de compra dos clientes e as características dos produtos mais vendidos.

## Análise Realizada:

  1. **Faturamento por Loja:** Calculamos o faturamento total de cada loja e visualizamos os resultados em um gráfico de barras para comparar o desempenho financeiro entre elas.

  2. **Vendas por Categoria de Produto:** Analisamos a distribuição das vendas por categoria de produto em todas as lojas combinadas, utilizando um gráfico de pizza para visualizar a proporção de cada categoria no faturamento total. Também geramos tabelas detalhadas de vendas por categoria e quantidade, faturamento por produto, e vendas por local e categoria.

  3. **Avaliação Média por Loja:** Calculamos a média das avaliações de compra para cada loja para entender a satisfação geral dos clientes.

  4. **Produtos Mais e Menos Vendidos por Loja:** Identificamos os produtos com o maior e menor número de vendas em cada loja para entender as preferências dos clientes e o desempenho individual dos produtos.

  5. **Frete Médio por Loja:** Calculamos o frete médio cobrado por cada loja e visualizamos essa informação em um gráfico de barras para comparar os custos de envio.


# Relatório de Vendas das Lojas

Este relatório apresenta uma explicação detalhada dos principais Análises obtidos a partir da análise dos dados de vendas das quatro lojas.

---

## Análise de Faturamento por Loja

Nossa análise inicial focou em determinar o desempenho financeiro de cada loja, calculando o faturamento total.

*   **Análise Principal:** A **Loja 1** demonstrou ser a loja com o maior faturamento absoluto, indicando um desempenho de vendas superior em comparação com as outras unidades.
*   **Observação:** Embora a Loja 1 tenha liderado, as diferenças de faturamento entre as lojas foram relativamente pequenas, com todas as unidades apresentando faturamentos na casa do milhão de reais.

---

## Análise de Vendas por Categoria de Produto

Em seguida, investigamos quais categorias de produtos impulsionam as vendas no geral, combinando os dados de todas as lojas.

*   **Análise Principal:** As categorias de **'eletronicos'** e **'eletrodomesticos'** são as que mais contribuem significativamente para o faturamento total, representando a maior fatia do "bolo" das vendas.
*   **Observação:** Isso sugere que produtos dessas categorias são os mais procurados pelos clientes no conjunto das lojas. Outras categorias como 'livros' e 'utilidades domesticas' tiveram uma participação menor no faturamento.

---

## Análise de Avaliação Média por Loja

Para entender a satisfação dos clientes, calculamos a média das avaliações de compra para cada loja.

*   **Análise Principal:** Todas as lojas recebem avaliações consistentemente positivas dos clientes, com médias acima de 3.9 em uma escala de 5.
*   **Observação:** Isso indica que, em geral, os clientes estão satisfeitos com suas compras em todas as unidades, o que é um bom indicador de qualidade de serviço e produto.

---

## Análise de Produtos Mais e Menos Vendidos por Loja

Identificamos os produtos com maior e menor volume de vendas em cada loja para entender as preferências locais e o desempenho individual dos produtos.

*   **Análise Principal:** Os produtos mais e menos vendidos variam de loja para loja, destacando as diferenças nas demandas e possivelmente nas estratégias de estoque ou promoções regionais.
*   **Observação:** Esta análise pontual pode ajudar cada loja a otimizar seu inventário e focar em produtos de alto desempenho, enquanto avalia o que pode ser feito para impulsionar as vendas dos produtos com menor saída.

---

## Análise de Frete Médio por Loja

Por fim, analisamos os custos médios de frete associados a cada loja.

*   **Análise Principal:** Existe uma variação no frete médio cobrado pelas lojas, com a **Loja 1** apresentando o valor mais alto e a **Loja 4** o mais baixo.
*   **Observação:** O custo do frete pode influenciar a decisão de compra do cliente. A diferença no frete médio pode ser um fator a ser considerado ao analisar a competitividade de cada loja, especialmente para clientes que valorizam o custo de envio.

---

**Conclusão:** A análise dos dados de vendas fornece uma visão sobre o desempenho das lojas, as preferências dos clientes por categorias e produtos, a satisfação geral com o serviço e as diferenças nos custos de frete.
